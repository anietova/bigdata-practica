```python
# https://pypi.org/project/websocket_client/
# https://finnhub.io/docs/api/websocket-trades
token="xxx"

import websocket



def on_message(ws, message):
    print(message)
    # Cerramos el websocket tras el primer mensaje
    ws.close()
    
def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

def on_open(ws):
    ws.send('{"type":"subscribe","symbol":"BINANCE:BTCUSDT"}')

if __name__ == "__main__":
    websocket.enableTrace(False)
    ws = websocket.WebSocketApp("wss://ws.finnhub.io?token="+token,
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)
    ws.on_open = on_open
    ws.run_forever()
```

```python
import json
import boto3
import websocket

KINESIS_STREAM = "mi-stream"

kinesis = boto3.client("kinesis", region_name="eu-west-1")

def on_message(ws, message):
    data = json.loads(message)

    kinesis.put_record(
        StreamName=KINESIS_STREAM,
        Data=json.dumps(data),
        PartitionKey="partition-1"
    )

def on_error(ws, error):
    print("Error:", error)

def on_close(ws):
    print("Conexi√≥n cerrada")

def on_open(ws):
    print("Conectado al WebSocket")

ws = websocket.WebSocketApp(
    "wss://endpoint.websocket.externo",
    on_message=on_message,
    on_error=on_error,
    on_close=on_close,
    on_open=on_open
)

ws.run_forever()
```
