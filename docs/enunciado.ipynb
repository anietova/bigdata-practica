# Actividad Cloud con AWS

Esta actividad consiste en que **implementéis un pequeño proyecto de análisis de datos utilizando algunas de las herramientas que nos ofrece Amazon Web Services (AWS)**. Se trata, pues, de una implementación libre de un proyecto en la nube de AWS.

Para hacerlo, utilizaremos una serie de recursos a los que os hemos dado acceso en la plataforma educativa AWS Academy, que son prácticamente los mismos que tenemos en la versión comercial/empresarial, pero sin que tengáis que preocuparos por daros de alta ni por los costos de su uso. A la hora de implementar el proyecto, os recomendamos maximizar el uso de los servicios PaaS y minimizar el uso de servicios IaaS. Por ejemplo, podéis utilizar los servicios S3, Lambda, Kinesis, EMR, EC2, Glue, Athena o Redshift, pero sois libres de utilizar los servicios adecuados para la recolección, almacenamiento, procesamiento y análisis de datos en función de vuestro problema. Podréis encontrar la lista de servicios disponibles dentro de la plataforma AWS Academy.

En cuanto a la evaluación, esta se realizará a partir de las rúbricas que encontraréis en este documento para cada uno de los apartados.

## AWS Academy, https://www.awsacademy.com/ 

Antes de empezar a desarrollar el proyecto, es muy recomendable realizar el curso **AWS Academy Cloud Foundations** al que os hemos dado acceso en la plataforma AWS Academy. Si superáis el presupuesto del laboratorio, se desactivará la cuenta del laboratorio y se perderá todo el progreso y los recursos.

El curso es bastante extenso, pero debéis leer y hacer las prácticas que hay en los siguientes módulos:

## **Ejercicio 3**: KPIs Específicos para un Proyecto de Análisis de Datos en AWS *(2 puntos)*

En un proyecto de análisis de datos desarrollado en AWS (Amazon Web Services), los KPIs no solo se centran en los resultados de la información (la precisión de los datos o los insights generados), sino también en el rendimiento, la eficiencia y el coste de la infraestructura en la nube.

A continuación se propone una estructura de KPIs clave, dividida por áreas:

### 1. KPIs de Coste y Eficiencia en AWS (FinOps)

* **KPI:** Coste Medio por Carga de Trabajo
    * **Descripción:** Coste total de AWS / Cantidad de datos procesados o insights generados.
    * **Objetivo del Proyecto:** Reducir el coste por unidad de análisis en un 10%.
* **KPI:** Tasa de Utilización de los Recursos
    * **Descripción:** Porcentaje de tiempo en que los recursos de AWS provisionados se están utilizando activamente.
    * **Objetivo del Proyecto:** Mantener por encima del 80%.
* **KPI:** Coste de Almacenamiento por GB
    * **Descripción:** Coste mensual para mantener los datos en el data lake (S3).
    * **Objetivo del Proyecto:** Optimizar el almacenamiento con políticas de retención adecuadas.

### 2. KPIs de Rendimiento y Velocidad

* **KPI:** Tiempo de Latencia de la Consulta
    * **Descripción:** Tiempo medio de respuesta de consultas críticas (en segundos).
    * **Objetivo del Proyecto:** Reducir la latencia de las consultas clave por debajo de 5 segundos.
* **KPI:** Tiempo de Ejecución del Pipeline ETL
    * **Descripción:** Tiempo total (en minutos) desde la ingesta de datos hasta su disponibilidad para el análisis (AWS Glue/Step Functions).
    * **Objetivo del Proyecto:** Reducir el tiempo de ejecución del pipeline ETL en un 15%.
* **KPI:** Tasa de Crecimiento de Datos
    * **Descripción:** GB/TB de datos nuevos ingeridos por día/mes.
    * **Objetivo del Proyecto:** Asegurar que el sistema (Kinesis/MSK) pueda escalar para gestionar este volumen.

### 3. KPIs de Calidad y Fiabilidad de los Datos

* **KPI:** Tasa de Error de Ingesta
    * **Descripción:** Porcentaje de registros o ficheros que fallan en la validación durante la carga a AWS S3.
    * **Objetivo del Proyecto:** Mantener la tasa de error de ingesta por debajo del 0,5%.
* **KPI:** Porcentaje de Inconsistencia de Datos
    * **Descripción:** Número de entradas no válidas / Total de entradas analizadas.
    * **Objetivo del Proyecto:** Eliminar o minimizar los datos inconsistentes.
* **KPI:** Disponibilidad del Servicio
    * **Descripción:** Tiempo que los servicios de análisis (Redshift, QuickSight) están operativos (en porcentaje).
    * **Objetivo del Proyecto:** Mantener la disponibilidad del servicio por encima del 99,9%.

### 4. KPIs de Impacto y Adopción

* **KPI:** Adopción de los Informes
    * **Descripción:** Usuarios activos diarios/semanales que acceden a los dashboards (QuickSight).
    * **Objetivo del Proyecto:** Incrementar el uso de los dashboards clave en un 20% en tres meses.
* **KPI:** Generación de Ingresos (o Ahorro) Atribuible
    * **Descripción:** Valor monetario generado o ahorrado gracias a los insights del proyecto.
    * **Objetivo del Proyecto:** Demostrar un ROI positivo para el proyecto.
* **KPI:** Precisión del Modelo Predictivo (si aplica)
    * **Descripción:** Métrica clave del modelo de Machine Learning (Puntuación F1, AUC, o Precisión).
    * **Objetivo del Proyecto:** Alcanzar una precisión del modelo del 90% o superior.
